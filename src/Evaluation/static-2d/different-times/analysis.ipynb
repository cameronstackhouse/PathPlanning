```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("dark")
```

```python
with open("evaluation_results.json") as json_data:
    data = json.load(json_data)
    json_data.close()
```

```python
data[1]["Success Rate"]
```

```python
res = {
    "8": data[0]["Success Rate"],
    "16": data[1]["Success Rate"],
    "32": data[2]["Success Rate"],
    "64": data[3]["Success Rate"],
}
```

```python
intervals = list(res.keys())
success_rates = list(res.values())

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(intervals, success_rates, marker='o', linestyle='-', color='b')

# Adding titles and labels
plt.title('Success Rate at Different Time Bounds')
plt.xlabel('Time (Seconds)')
plt.ylabel('Success Rate')

# Display the plot
plt.grid(True)
plt.show()
```
